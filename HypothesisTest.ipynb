```python
mean_diff = df.loc[df['Fertilizer'] == False,'Growth Duration'].mean() - df.loc[df['Fertilizer'] == True,'Growth Duration'].mean()
mean_diff
```

```python
def permDF(dataFrame):
    return dataFrame.sample(frac = 1).reset_index(drop = True)
permDF(df)
```

```python
test_stat = np.array([])
reps = 10000
for i in range(reps):
    permuted_df = pd.DataFrame({'Fertilizer' : df['Fertilizer'], 'Growth Duration' : permDF(df['Growth Duration'])})
    test_stat = np.append(test_stat, permuted_df.loc[permuted_df['Fertilizer'] == False,'Growth Duration'].mean() - permuted_df.loc[permuted_df['Fertilizer'] == True,'Growth Duration'].mean())
```

```python
# test_stat.max()
mean_diff
```

```python
p_value = np.count_nonzero(test_stat <= mean_diff) / reps
p_value
```
