```python
import pandas as pd
from zipfile import ZipFile
```

```python
zip_file = ZipFile('us-baby-names.zip')
```

```python
df = pd.read_csv(zip_file.open('NationalNames.csv'))
```

```python
df.head()
```

```python
df.loc[df['Year'] == 2014, :].sort_values('Count', ascending = False)[:5]
```

```python
df.loc[df['Year'] == 2014, :].sort_values('Count', ascending = False)[:5]
```

```python
statewise_df = pd.read_csv(zip_file.open('StateNames.csv'))
```

```python
states_df_CA = statewise_df.loc[statewise_df['State'] == 'CA',:]
states_df_CA.loc[states_df_CA['Year'] == 2014,:].sort_values('Count', ascending = False)[:5]
```
